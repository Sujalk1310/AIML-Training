```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Mental Health Chatbot

inter: 2

intra: 6

KMP_BLOCKTIME: 1

Test_Set: 20

```python
df = pd.read_csv("./benchmarking.csv")
Model = "Mental Health Chatbot"

data1 = df.loc[(df['Model'] == Model) & (df["benchmark_parameter"] == "Inference_Speed"), ["With_ONEAPI", "WITHOUT_ONEAPI"]].values.flatten()
data2 = df.loc[(df['Model'] == Model) & (df["benchmark_parameter"] == "Latency"), ["With_ONEAPI", "WITHOUT_ONEAPI"]].values.flatten()
data3 = df.loc[(df['Model'] == Model) & (df["benchmark_parameter"] == "Throughput"), ["With_ONEAPI", "WITHOUT_ONEAPI"]].values.flatten()
data4 = df.loc[(df['Model'] == Model) & (df["benchmark_parameter"] == "Training_Time"), ["With_ONEAPI", "WITHOUT_ONEAPI"]].values.flatten()
```

```python
colors = ['Blue', 'Orange']

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

bars1 = axes[0, 0].bar(range(len(data1)), data1, color=colors)
axes[0, 0].set_title("Inference Speed")
axes[0, 0].text(0.5, 1.11, '(Higher is better)', ha='center', va='center', transform=axes[0, 0].transAxes)

bars2 = axes[0, 1].bar(range(len(data2)), data2, color=colors)
axes[0, 1].set_title("Latency")
axes[0, 1].text(0.5, 1.11, '(Lower is better)', ha='center', va='center', transform=axes[0, 1].transAxes)

bars3 = axes[1, 0].bar(range(len(data3)), data3, color=colors)
axes[1, 0].set_title("Throughput")
axes[1, 0].text(0.5, 1.11, '(Higher is better)', ha='center', va='center', transform=axes[1, 0].transAxes)

bars4 = axes[1, 1].bar(range(len(data4)), data4, color=colors)
axes[1, 1].set_title("Training Time")
axes[1, 1].text(0.5, 1.11, '(Lower is better)', ha='center', va='center', transform=axes[1, 1].transAxes)

axes[0, 0].legend(bars1, ['Intel oneAPI', 'Core CPU'])
axes[0, 1].legend(bars2, ['Intel oneAPI', 'Core CPU'])
axes[1, 0].legend(bars3, ['Intel oneAPI', 'Core CPU'])
axes[1, 1].legend(bars4, ['Intel oneAPI', 'Core CPU'])

plt.suptitle(Model)

plt.tight_layout()

plt.show()
```

```python
print("Inference Time Rate: {}".format(data1[0]/data1[1]))
print("Latency Rate: {}".format(data2[0]/data2[1]))
print("Throughput Rate: {}".format(data3[0]/data3[1]))
print("Training Time Rate: {}".format(data4[0]/data4[1]))
```
