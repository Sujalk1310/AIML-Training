```python
def DFT(x):
  N=len(x)
  n=np.arange(N)
  k=n.reshape((N,1))
  e=np.exp(-2j*np.pi*k*n/N)
  X=np.dot(e,x) # not 2 its e #omkay
  return X
```

```python
X=DFT(x)
N=len(X)
n=np.arange(N)
T=N/s
freq=n/T
```

```python
plt.stem(freq, abs(X), 'b' , markerfmt="_", basefmt="_b")
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.show()
```
