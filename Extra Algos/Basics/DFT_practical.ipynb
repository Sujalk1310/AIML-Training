```python
import numpy as np
import matplotlib.pyplot as plt
a = 13
f = 1
phi = 0
s = 100
tr = 1./100
t = np.arange(0, 1, tr)
x = a * np.sin(2 * np.pi * f * t + phi)
plt.plot(t, x, 'b')
plt.show()
```

```python
def DFT(x):
  N = len(x)
  n = np.arange(N)
  K = n.reshape((N,1))
  e = np.exp(-2j * np.pi * K * n/N)
  X = np.dot(e, x)
  return X
```

```python
X = DFT(x)
N = len(x)
n = np.arange(N)
T = N/s
freq = n/T
plt.stem(freq, abs(X), 'b',markerfmt = " ", basefmt = "_b")
plt.show()
```
