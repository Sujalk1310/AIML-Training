```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 1: Load and describe the data
# Assuming you have a CSV file containing your dataset named 'house_prices.csv'
df = pd.read_csv('Dataset\HousingData.csv')
```

```python
# You can describe the data to get an overview of the features
print(df.describe())
```

```python
# Step 2: Build the correlation matrix
correlation_matrix = df.corr()
```

```python
# Step 3: Plot a heatmap of the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Step 4: Keep only the features with correlation > abs(0.5) with 'price'
threshold = 0.5  # You can adjust this threshold as needed
corr_with_price = correlation_matrix['MEDV'].abs()
selected_features = corr_with_price[corr_with_price > threshold].index.tolist()
```

```python
# Now, 'selected_features' contains the names of features with correlation > abs(0.5) with 'price'
# You can use these features for your modeling.
print("Selected Features:", selected_features)
```

```python
# Create a correlation matrix for selected features
selected_corr_matrix = df[selected_features].corr()

# Plot a heatmap for the selected features
plt.figure(figsize=(10, 6))
sns.heatmap(selected_corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap for Selected Features')
plt.show()
```
