# 1. Find the missing values of hours and distance and replace with the mean and find the correlation and visualize it. And find the correlation by normalizing the data using decimal scaling technique. And find the difference in both correlations. 


```python
import pandas as pd
import seaborn as sns
# Load the dataset into a DataFrame
df = pd.read_csv('./Dataset/flights.csv')

# Check for missing values
missing_hours = df['hour'].isnull()
missing_distance = df['distance'].isnull()

# Replace missing values with mean
df['hour'].fillna(df['hour'].mean(), inplace=True)
df['distance'].fillna(df['distance'].mean(), inplace=True)
```

```python
df
```

```python
# Decimal Scaling normalization
df['hours_decimal'] = df['hour'] / 10
df['distance_decimal'] = df['distance'] / 10

# Calculate correlation
decimal_correlation = df['hours_decimal'].corr(df['distance_decimal'])

# Visualize the data
import matplotlib.pyplot as plt

plt.scatter(df['hours_decimal'], df['distance_decimal'])
plt.xlabel('Normalized Hours (Decimal Scaling)')
plt.ylabel('Normalized Distance (Decimal Scaling)')
plt.title(f'Correlation: {decimal_correlation}')
plt.show()

print(f'Correlation with Decimal Scaling: {decimal_correlation}')
```

```python
# Calculate the correlation matrix for the normalized columns
correlation_matrix_decimal = df[['hours_decimal', 'distance_decimal']].corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_decimal, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap (Decimal Scaling)')
plt.show()
```

# 2. Find the missing values of hours and distance and replace with the median and find the correlation and visualize it. And find the correlation by normalizing the data using min-max normalization technique. And find the difference in both correlation. 

```python
# Load the dataset into a DataFrame
df = pd.read_csv('./Dataset/flights.csv')

# Check for missing values
missing_hours = df['hour'].isnull()
missing_distance = df['distance'].isnull()

# Replace missing values with mean
df['hour'].fillna(df['hour'].median(), inplace=True)
df['distance'].fillna(df['distance'].median(), inplace=True)
```

```python
df
```

```python
# Min-Max normalization
df['hours_minmax'] = (df['hour'] - df['hour'].min()) / (df['hour'].max() - df['hour'].min())
df['distance_minmax'] = (df['distance'] - df['distance'].min()) / (df['distance'].max() - df['distance'].min())

# Calculate correlation
minmax_correlation = df['hours_minmax'].corr(df['distance_minmax'])

# Visualize the data
plt.scatter(df['hours_minmax'], df['distance_minmax'])
plt.xlabel('Normalized Hours (Min-Max Scaling)')
plt.ylabel('Normalized Distance (Min-Max Scaling)')
plt.title(f'Correlation: {minmax_correlation}')
plt.show()

print(f'Correlation with Min-Max Scaling: {minmax_correlation}')
```

```python
# Calculate the correlation matrix for the normalized columns
correlation_matrix_minmax = df[['hours_minmax', 'distance_minmax']].corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_minmax, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap (MinMax Scaling)')
plt.show()
```

# 3. Find the missing values of hours and distance and replace with the mode and find the correlation and visualize it. And find the correlation by normalizing the data using z- score normalization technique. And find the difference in both correlations.

```python
# Load the dataset into a DataFrame
df = pd.read_csv('./Dataset/flights.csv')

# Check for missing values
missing_hours = df['hour'].isnull()
missing_distance = df['distance'].isnull()

# Replace missing values with mean
df['hour'].fillna(df['hour'].mode()[0], inplace=True)
df['distance'].fillna(df['distance'].mode()[0], inplace=True)
```

```python
df
```

```python
# Z-Score normalization
df['hours_zscore'] = (df['hour'] - df['hour'].mean()) / df['hour'].std()
df['distance_zscore'] = (df['distance'] - df['distance'].mean()) / df['distance'].std()

# Calculate correlation
zscore_correlation = df['hours_zscore'].corr(df['distance_zscore'])

# Visualize the data
plt.scatter(df['hours_zscore'], df['distance_zscore'])
plt.xlabel('Normalized Hours (Z-Score Scaling)')
plt.ylabel('Normalized Distance (Z-Score Scaling)')
plt.title(f'Correlation: {zscore_correlation}')
plt.show()

print(f'Correlation with Z-Score Scaling: {zscore_correlation}')
```

```python
# Calculate the correlation matrix for the normalized columns
correlation_matrix_zscore = df[['hours_zscore', 'distance_zscore']].corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_zscore, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap (Z-Score)')
plt.show()
```

# 4. At which mean/median/mode do you find the maximum correlation.

```python
correlations = {
    'Mean-Decimal': decimal_correlation,
    'Median-MinMax': minmax_correlation,
    'Mode-Z_Score': zscore_correlation
}

max_correlation_method = max(correlations, key=correlations.get)
max_correlation_value = correlations[max_correlation_method]

print(f'Maximum Correlation ({max_correlation_method}): {max_correlation_value}')
```
