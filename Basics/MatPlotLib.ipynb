# **Importing the Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60]
y = [1,2,5,8,3,7]
plt.plot(x, y, marker='o')
plt.show
```

```python
X = np.array([1, 8])
Y = np.array([3, 10])
plt.plot(X, Y)
plt.show()
```

```python
# Plotting without line
plt.plot(X, Y, 'o')
plt.show()
```

```python
X = np.array([1, 2, 6, 8])
Y = np.array([3, 8, 1, 10])
plt.plot(X, Y)
plt.show()
```

```python
Y = np.array([3, 8, 1, 10, 5, 7])
plt.plot(Y)
plt.show()
```

```python
Y = np.array([3, 8, 1, 10])
plt.plot(Y, marker='o')
plt.show()
```

```python
plt.plot(Y, marker='*')
plt.xlabel('Price')
plt.ylabel('Tenure')
plt.title('Sales')
plt.suptitle('Arduino')
plt.show()
```

```python
# Labels and Titles
X = np.array([80, 85, 90,95, 100, 105, 110, 115, 120, 125])
Y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(X, Y)
plt.show()
```

```python
Y = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1)
plt.plot(Y)
Y = np.array([0, 1, 2, 4])
plt.subplot(1,2,2)
plt.plot(Y)
plt.show()
```

```python
Y = np.array([3, 8, 1, 10])
plt.subplot(2, 1, 1)
plt.plot(Y)
Y = np.array([0, 1, 2, 4])
plt.subplot(2,1,2)
plt.plot(Y, color='black', marker="*")
plt.show()
```

```python
Y = np.array([3, 8, 1, 10])
plt.subplot(2, 3, 1)
plt.plot(Y)
Y = np.array([10, 20, 35, 40])
plt.subplot(2, 3, 2)
plt.plot(Y)
Y = np.array([5, 36, 15, 20])
plt.subplot(2, 3, 3)
plt.plot(Y)
Y = np.array([8, 12, 24, 42])
plt.subplot(2, 3, 4)
plt.plot(Y)
Y = np.array([5, 11, 5, 25])
plt.subplot(2, 3, 5)
plt.plot(Y)
Y = np.array([5, 6, 10, 22])
plt.subplot(2, 3, 6)
plt.plot(Y)
plt.show()
```

```python
# We need to define X Cords in  Scatter()
X = np.array([1, 2, 3, 4])
Y = np.array([3, 8, 1, 10])
plt.scatter(X, Y, marker='o')
plt.show()
```

```python
X = np.array([1, 2, 3, 4])
Y = np.array([3, 8, 1, 10])
plt.bar(X, Y, color="orange")
plt.show()
```

```python
Y = np.array([3, 8, 1, 10])
plt.pie(Y, labels=['1st', '2nd', '3rd', '4th'])
plt.show()
```
