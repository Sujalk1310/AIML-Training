# DFT

```python
def DFT(x):
    N = len(x)
    n = np.arange(N)
    K = n.reshape((N, 1))
    e = np.exp(-2j*np.pi*K*n/N)
    X = np.dot(e, x)
    return X
```

```python
X = DFT(x)
N = len(x)
n = np.arange(N)
T = N/s
freq = n/T
plt.stem(freq, abs(X), 'b', markerfmt = ' ', basefmt = '_b')
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.show()
```
