```python
import numpy as np
numq=16
currentsol=[0 for x in range(numq)]
solution=[]
```

```python
currentsol
```

```python
def IsSafe(testRow,testCol):
          if  testRow==0:
            return True
          for row in range(0,testRow):
            if testCol==currentsol[row]:
              return False
            if abs(testRow-row)==abs(testCol-currentsol[row]):
              return False
          return True
```

```python
for Q0 in range(numq):
  if not IsSafe(0,Q0):
    continue
  else:
    currentsol[0]=Q0
    for Q1 in range(numq):
      if not IsSafe(1,Q1):
        continue
      else:
        currentsol[1]=Q1
        for Q2 in range(numq):
          if not IsSafe(2,Q2):
            continue
          else:
            currentsol[2]=Q2
            for Q3 in range(numq):
              if not IsSafe(3,Q3):
                continue
              else:
                currentsol[3]=Q3
                for Q4 in range(numq):
                  if not IsSafe(4,Q4):
                    continue
                  else:
                    currentsol[4]=Q4
                    for Q5 in range(numq):
                      if not IsSafe(5,Q5):
                        continue
                      else:
                        currentsol[5]=Q5
                        for Q6 in range(numq):
                            if not IsSafe(6,Q6):
                              continue
                            else:
                              currentsol[6]=Q6
                              for Q7 in range(numq):
                                  if not IsSafe(7,Q7):
                                    continue
                                  else:
                                    currentsol[7]=Q7
                                    solution.append(currentsol.copy())
```
