```python
s = 100 # Number of Samples
tr = 1.0/100  # Time Interval
t = np.arange(0,1,tr) # Generating values of t
a = 0.5 # Amplitude 
f = 7  # Frequency
Phi = 0 
x += a * np.sin(2*np.pi*f*t+Phi)
plt.plot(t,x,'b')
plt.style.use('seaborn-poster')
plt.xlabel("Location or time")
plt.ylabel("Amplitude")
plt.show()
```

```python
def DFT(x):
  N = len(x)
  n = np.arange(N)
  K = n.reshape((N,1))
  e = np.exp(-2j * np.pi * K * n /N)
  X = np.dot(e,x)
  # X --> Fourier Tranformation Value
  return X
```

```python
X = DFT(x)
N = len(x)
n = np.arange(N)
T = N/s
freq = n/T
plt.stem(freq,abs(X), 'b', markerfmt = " ", basefmt = "_b")
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.show()
```

# FFT

```python
from scipy.fftpack import fft,ifft
X = fft(x)
plt.stem(freq,np.abs(X), 'b', markerfmt = " ", basefmt = "_b")
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.show()
```
